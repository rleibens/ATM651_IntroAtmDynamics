### Background
**Kinematics** is not dynamics, it is just a *description* of how 2D vector fields vary in 2D. 

* Assume LINEAR dependence of u,v on x,y 
    * This involves 4 coefficients $c_1, c_2, c_3, c_4$
* Also, add a constant (not variying with x,y) vector field U,V 

$$ u = U + c_1 x  +c_2 y$$
$$ v = V + c_3 x  +c_4 y$$

## Exercise 1 Answer:
* $ div = \frac{\partial u}{\partial x} + \frac{\partial v}{\partial y}= \frac{\partial U}{\partial x} + \frac{\partial c_1x}{\partial x} + \frac{\partial c_2y}{\partial x} + \frac{\partial V}{\partial y} + \frac{\partial c_3x}{\partial y} + \frac{\partial c_4y}{\partial y}= 0 + c_1 + 0 + 0 + 0 + c_4 = c_1 + c_4$ 
* $ \zeta_{rel} = \frac{\partial v}{\partial x} - \frac{\partial u}{\partial y} = \frac{\partial V}{\partial x} + \frac{\partial c_3x}{\partial x} + \frac{\partial c_4y}{\partial x} - \frac{\partial U}{\partial y} - \frac{\partial c_1x}{\partial y} - \frac{\partial c_2y}{\partial y} = 0 + c_3 + 0 + 0 + 0 - c_2 = c_3 - c_2$
* $ def1 = \frac{\partial u}{\partial x} - \frac{\partial v}{\partial y} = \frac{\partial U}{\partial x} + \frac{\partial c_1x}{\partial x} + \frac{\partial c_2y}{\partial x} - \frac{\partial V}{\partial y} - \frac{\partial c_3x}{\partial y} - \frac{\partial c_4y}{\partial y} = 0 + c_1 + 0 + 0 + 0 - c_4 = c_1 - c_4$
* $ def2 = \frac{\partial v}{\partial x} + \frac{\partial u}{\partial y} = \frac{\partial V}{\partial x} + \frac{\partial c_3x}{\partial x} + \frac{\partial c_4y}{\partial x} + \frac{\partial U}{\partial y} + \frac{\partial c_1x}{\partial y} + \frac{\partial c_2y}{\partial y} = 0 + c_3 + 0 + 0 +0 + c_2 = c_3 + c_2$
